```python
# Import necessary libraries for text preprocessing
import string
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
# remove stopwords, punctuation, and normalize the corpus
stop = set(stopwords.words('english')) | {"paper", "study", "data"}
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()
```

1.2 Preprocessing

```python
import nltk
nltk.download('punkt_tab')

# Preprocess the text data:
def preprocess_text(doc):
    # Tokenization and lowercase
    tokens = nltk.word_tokenize(doc.lower())  # Robust tokenization with nltk

    # Remove stop words
    stop_free = [token for token in tokens if token not in stop]

    # Remove punctuation
    punc_free = [token for token in stop_free if token not in exclude]

    # Lemmatization
    normalized = [lemma.lemmatize(token) for token in punc_free]

    return normalized

# Preprocess the corpus
clean_corpus = [preprocess_text(doc) for doc in corpus]
```

```python
clean_corpus
```

```python
## Create a dictionary and corpus for Gensim

from gensim.corpora.dictionary import Dictionary

gensim_dict = Dictionary(clean_corpus)
gensim_corpus = [gensim_dict.doc2bow(text) for text in clean_corpus]
```

# Task 2: Building a Topic Model

2.1 Define a function to train an LDA model

```python
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np
from gensim.models import CoherenceModel
from gensim.corpora.dictionary import Dictionary

def train_lda_model(tfidf_matrix, num_topics, random_state=42):
    lda_model = LatentDirichletAllocation(n_components=num_topics, random_state=random_state)
    lda_model.fit(tfidf_matrix)
    return lda_model
```

```python
#extract topics
def extract_topics(lda_model, vectorizer, num_words=10):
    terms = vectorizer.get_feature_names_out()
    topics = [
        [terms[i] for i in topic.argsort()[-num_words:][::-1]]
        for topic in lda_model.components_
    ]
    return topics
```

2.2 Hyperparameter tuning using coherence scores

```python
from sklearn.feature_extraction.text import TfidfVectorizer

coherence_scores = []
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
tfidf_matrix = vectorizer.fit_transform([" ".join(doc) for doc in clean_corpus])
```

```python
#Tested for 1-20 topics and saw improvement for 5â€“9 Topics and sharp increase beyond 10 topics. So, I choosed 5 to 20 topics at first and then moved to 8â€“12 and 12-15 for more granularity 
```

```python
#Choose an appropriate number of topics by testing various values using cohenrence scores
for num_topics in range(12,15): 
    lda_model = train_lda_model(tfidf_matrix, num_topics)
    topics = extract_topics(lda_model, vectorizer)
    coherence_model = CoherenceModel(
        topics=topics, texts=clean_corpus, dictionary=gensim_dict, coherence='c_v'
    )
    coherence = coherence_model.get_coherence()
    coherence_scores.append((num_topics, coherence))
    print(f"Num Topics: {num_topics}, Coherence: {coherence:.4f}")
```

```python
# Select the best number of topics
best_num_topics = max(coherence_scores, key=lambda x: x[1])[0]
print(f"\nOptimal number of topics: {best_num_topics}")
```

2.3 Model Fitting:

```python
# Train final LDA model
final_lda_model = train_lda_model(tfidf_matrix, best_num_topics)
final_topics = extract_topics(final_lda_model, vectorizer)
```

```python
# Display topics
print("\nTopics:")
for i, topic in enumerate(final_topics):
    print(f"Topic {i+1}: {', '.join(topic)}")
```

# Task 3: Visualizing and Interpreting the results

3.1 Visualization

```python
#Using PyLDAvis 
import pyLDAvis

# Calculate the required parameters for PyLDAvis
topic_term_dists = final_lda_model.components_ / final_lda_model.components_.sum(axis=1)[:, np.newaxis]
doc_topic_dists = final_lda_model.transform(tfidf_matrix)
vocab = vectorizer.get_feature_names_out()
term_frequency = tfidf_matrix.sum(axis=0).A1
doc_lengths = [len(doc) for doc in clean_corpus]  # Number of tokens in each document

# Prepare the visualization
visualization = pyLDAvis.prepare(
    topic_term_dists=topic_term_dists,
    doc_topic_dists=doc_topic_dists,
    doc_lengths=doc_lengths,
    vocab=vocab,
    term_frequency=term_frequency
)

# Display visualization
pyLDAvis.display(visualization)
```

Interpretation: The PyLDAvis visualizations show well-separated clusters for some topics (e.g., Topic 1 and Topic 2), while others like Topics 11 and 12 are closely aligned, which reflects the shared thematic elements in those areas. Topics 6, 8, and 11 appear closely clustered indicating overlapping terms or themes, particularly in protein-related research. Topics like 3 and 10 are moderately isolated, representing unique research areas such as robotics and customer sentiment.

```python
#Using WordCloud

from wordcloud import WordCloud
import matplotlib.pyplot as plt

for i, topic in enumerate(final_topics):
    plt.figure()
    wordcloud = WordCloud(width=800, height=400, background_color='white')
    wordcloud.generate_from_frequencies(dict(zip(topic, final_lda_model.components_[i])))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Topic {i+1}")
    plt.show()
```

```python
#Topic Prevalence
import seaborn as sns

# heatmap of topic contributions per document
doc_topic_distribution = final_lda_model.transform(tfidf_matrix)
plt.figure(figsize=(12, 8))
sns.heatmap(doc_topic_distribution, cmap="YlGnBu", annot=False)
plt.title("Topic Contribution Per Document")
plt.xlabel("Topics")
plt.ylabel("Documents")
plt.show()
```

Interpretation: Most documents are strongly aligned with a single topic, suggesting the topics are distinct and meaningful. Topics that appear in fewer documents might represent niche themes or redundant content. Some topics (e.g., Topic 5 and Topic 8) appear less frequently across documents.

```python
#Using bar chart
import matplotlib.pyplot as plt

topic_distribution = np.sum(final_lda_model.transform(tfidf_matrix), axis=0)  # Sum probabilities across all documents
plt.bar(range(1, best_num_topics + 1), topic_distribution)  # Use best_num_topics
plt.xlabel('Topic Number')
plt.ylabel('Document Share')
plt.title('Topic Prevalence')
plt.show()
```

Interpretation:
Topic 6 has the highest prevalence, appearing in the largest number of documents. Topic 7 has the lowest prevalence, suggesting it is rarely the dominant topic in documents. The remaining topics (e.g., Topics 2, 3, 10) have similar prevalence, which reflects a balanced contribution of themes.

```python
# Add a pairwise similarity plot between topics
topic_similarity = np.corrcoef(doc_topic_distribution.T)
plt.figure(figsize=(10, 8))
sns.heatmap(topic_similarity, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Topic Similarity Matrix")
plt.xlabel("Topics")
plt.ylabel("Topics")
plt.show()
```

Interpretation: Topics 6 and 8 show a significant negative correlation (-0.62), suggesting contrasting themes. This indicates that the topics are distinct and may represent opposing perspectives or contexts. Topics 10 and 11 have a moderate positive correlation (0.31), which might indicate some overlap or shared themes between these two topics. There are no strongly overlapping topics, meaning your model has done well in separating themes.
